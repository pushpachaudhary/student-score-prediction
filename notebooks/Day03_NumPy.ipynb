```python
import numpy as np
```

```python
print("Numpy Version:", np.__version__)
```

```python
#1D Array
arr1 = np.array([10,20,30,40,50])

#2D Array
arr2 = np.array([
    [1,2,3],
    [4,5,6]
])

print(f"1D Array: {arr1}")
print(f"2D Array: {arr2}")
```

```python
#Array properties
print("Shape : ",arr2.shape)
print("size : ",arr2.size)
print("Dimension : ",arr2.ndim)
print("DAta type : ",arr2.dtype)
```

```python
#Special Arrays
zeros = np.zeros((3,3))
print("Zeros Array:\n",zeros)

ones = np.ones((2,4))
print("Ones Array:\n",ones)

identity = np.eye(4)
print("Identity Matrix:\n",identity)
```

```python
#Array Indexing and Slicing
numbers = np.array([10,20,30,40,50,60,70,80])
print("Original Array:", numbers)

print("\nElement at index 1:", numbers[1])
print("\nElement at index -1:", numbers[-1])
print("\nElement at index 5:", numbers[5])

print("\nElement at index 2 to 5:", numbers[2:5])
print("\nElement at index 0 to 5 with step 2:", numbers[0:5:2])
print("\nElement at index 3 to end:", numbers[3:])
print("\nElement at index 0 to 8 in reverse:", numbers[::-1])

```

```python
#Arithematic Operations

a = np.array([10,20,30])
b = np.array([2,4,6])

print("Addition:", a+b)
print("Subtraction:", a-b)
print("Multiplication:", a*b)
print("Division:", a/b)
```

```python
#Mathematical Functions

num = np.array([4,9,16,25,36])

print("Square Root :",np.sqrt(num))
print("square :",np.square(num))
print("Exponential :",np.exp(num))
print("Logarithm :",np.log(num))


```

```python
#Statistical Functions

marks = np.array([85, 92, 78, 90, 88])
print("mean :", np.mean(marks))
print("median :", np.median(marks))
print("standard deviation :", np.std(marks))
print("maximum :", np.max(marks))
```

```python
#Reshaping arrays

arr = np.array([1, 2, 3, 4, 5])

print("original array:", arr)
print("Reshaped array :\n", arr.reshape(5,1))
```

```python
#Random Numbers

np.random.seed(42)
random_numbers = np.random.randint(1,100,size = 10)
print("Random numbers from 1 to 100:", random_numbers)
```

```python
#Mini Practice Problem (Student Scores)
scores = np.array([78, 85, 92, 88, 76, 95, 89, 84, 91, 87])

print("Scores:", scores)
print("Average score:", np.mean(scores))
print("Highest score:", np.max(scores))
print("Lowest score:", np.min(scores))

passed  = scores[scores >= 60]
print("Students scoring above 60:", passed)
print("Number of students scoring above 60:", len(passed))
```

```python
#Challenge problem

study_hrs = np.array([2,3,4,5,6])
scores = np.array([70, 75, 80, 85, 90])

improvement = scores - scores[0]

print("study hours : ", study_hrs)
print("scores : ", scores)
print("Improvement : ", improvement)
print("Average Scores : ", np.mean(scores))
```
