```python
def table(n):
    for i in range(1,11):
        print(f"{n} * {i} = {n*i}")

table(5)     
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

factorial(10)
```

```python
def sum_of_natural_numbers(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    print(f"sum of first {n} natural numbers = {sum}")

        


sum_of_natural_numbers(200)
```

```python
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        print(a ,end = " ")
        a, b = b, a + b
fibonacci(20)
```
