```python
import pandas as pd 
import numpy as np 

print("Pandas Version : ", pd.__version__)
```

```python
#creating a sample dataset
student_data = {
    'student_name': ['Alice', 'Bob', 'Charlie', 'David','Aman', 'karan', 'Ankita','prince','Priya','Anjali','Anita','Aniket','Meera','Hina','Palak','Harshita', 'Rahul','Rohit', 'Preeti', 'Chahana', 'Sohan'],
    'study_hours': [5,7, 6, 7, 7, 5,  6, 6, 4,   4, 3, 2,   4 , 8, 3, 5, 5, 2, 2, 8, 5,],
    'Score': [85, 95, 80, 95, 77, 65, 99, 95, 60, 67,57, 60, 73,80,48,40,75,46, 49, 88, 78]
}
df = pd.DataFrame(student_data)
df
```

```python
#Display First and Last Records of the dataset
print("First 4 rows of the dataset : \n", df.head(4))
print("\nLast 4 rows of the dataset : \n", df.tail(4))
```

```python
#Student scoring above 50
high_scores = df[df['Score'] >= 70]
high_scores
```

```python
#sorting of students according to scores
df.sort_values(by = 'Score', ascending = False)
```

```python
#Adding new column of result
df["Result"] = np.where(df["Score"] >= 60 , "PASS" ,"FAIL")
df
```

```python
#Average score
average = df['Score'].mean()
print("Average Score : ",average)
```

```python
#Highest and Lowest marks
print("Highest Score : ",df["Score"].max())
print("Lowest Score : " ,df["Score"].min())
```

```python
#Top performer
top_student = df.loc[df["Score"].idxmax()]
top_student
```

```python
#Total no. of students passed and failed
df["Result"].value_counts()
```
